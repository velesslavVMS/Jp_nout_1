#### 2. Найдите выручку компании в зависимости от месяца. 
Напишите функцию, которая на вход принимает списки с датами и с выручкой, а на выходе вернет словарь, где ключи - это месяцы, а значения - это выручка.

```python
dates = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05', '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29', '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18', '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02', '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01', '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09', '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13', '2021-10-26', '2021-12-09']
revenues = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037, 4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]
def rez(date, revenue):
   date1 = [i[5:7] for i in date]  #list comprehensions

   sum09 = 0
   sum10 = 0
   sum11 = 0
   sum12 = 0
   for j in range(len(revenue)):
    
      if date1[j] == '09':
         sum09 += revenue[j]
      elif date1[j] == '10':
         sum10 += revenue[j]
      elif date1[j] == '11':
         sum11 += revenue[j]
      elif date1[j] == '12':
         sum12 += revenue[j]
   total = [sum09, sum10, sum11, sum12]
   month = ['09','10','11','12']
   return dict(zip(month, total))

print(rez(dates, revenues))
```
